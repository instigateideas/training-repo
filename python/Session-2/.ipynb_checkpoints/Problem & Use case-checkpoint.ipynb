### 3. What’s the impact of the toss on each team?

````python
matches_df = pd.read_csv("data/ipl/IPL Matches 2008-2020.csv")
matches_df.head()
````

````python
# impact of toss leading to win
toss_winner_win = matches_df.loc[matches_df.toss_winner == matches_df.winner].groupby("toss_winner").agg({"winner": 'count'})
toss_winner_lose = matches_df.loc[matches_df.toss_winner != matches_df.winner].groupby("toss_winner").agg({"winner": 'count'})
````

````python
toss_winner_win = matches_df.loc[(matches_df.toss_winner == matches_df.winner) & ((matches_df.toss_decision == "bat") )].groupby("toss_winner").agg({"winner": 'count'})
toss_winner_lose = matches_df.loc[(matches_df.toss_winner == matches_df.winner) & (matches_df.toss_decision == "field")].groupby("toss_winner").agg({"winner": 'count'})
````

````python
toss_winner_win
````

````python
toss_winner_lose
````

````python
toss_data = pd.merge(toss_winner_win, toss_winner_lose, left_index=True, right_index=True).reset_index()
toss_data.columns = ["Toss Winner", "Toss Win & Bat", "Toss Win  & Field"]
````

````python
toss_data["Total toss oppertunity"] = toss_data[["Toss Win & Bat", "Toss Win  & Field"]].sum(axis=1)
````

````python
toss_data
````

````python
toss_data["Toss Win  & Field"]
````

````python
toss_data["Percentage of success_field"] = round(toss_data["Toss Win  & Field"]/toss_data["Total toss oppertunity"]*100, 2)
````

````python
# Below table gives the max success in a match - Toss impact
toss_data.sort_values(by="Percentage of success", ascending=False)
````

## 7. Vizualize the % of the win for each team - pie chart?

````python
matches_df.columns
````

````python
matches_df.head()
````

````python
total_matches = len(matches_df)
total_matches
````

````python
team1_cnt = matches_df.groupby("team1").agg({"team1": 'count'})
team2_cnt = matches_df.groupby("team2").agg({"team2": 'count'})
team_matches = pd.merge(team1_cnt, team2_cnt, left_index=True, right_index=True)
team_matches["Total"] = team_matches.sum(axis=1)
````

````python
team_matches
````

````python
winner_cnt = matches_df.groupby("winner").agg({"winner": 'count'})
````

````python
winner_cnt
````

````python
winner_pct_df = pd.merge(team_matches, winner_cnt, left_index=True, right_index=True)
````

````python
winner_pct_df
````

````python
winner_pct_df["win_pct"] = round(winner_pct_df["winner"]/winner_pct_df["Total"]*100, 2)
````

````python
winner_pct_df
````

````python
win_pct_sorted = winner_pct_df.sort_values(by="win_pct", ascending=False)
````

````python
win_pct_sorted.columns
````

````python
win_pct_sorted
````

````python
win_pct_sorted["lost_pct"] = 100 - win_pct_sorted["win_pct"]
````

````python
win_pct_sorted
````

````python
for i in win_pct_sorted.iterrows():
    win_pct = i[1]["win_pct"]
    loss_pct = i[1]["lost_pct"]
    team = i[1]["team1"]
    graph_data = [win_pct, loss_pct]
    plt.pie(graph_data)
    plt.title(team)
    plt.show()
    
````

````python
plt.pie([62.5, 37.5])
plt.xlabel("loss_pct")
plt.ylabel("win_pct")
plt.show()
````

````python
win_pct_sorted.drop(["team1", "team2"], axis=1, inplace=True)
````

````python
win_pct_sorted.reset_index(inplace=True)
````

````python
plt.bar("team1", "win_pct", data=win_pct_sorted)
plt.title("Win Percentage - Teamwise")
plt.xlabel("Team")
plt.ylabel("Win Percentage")
plt.xticks(rotation=90)
plt.show()
````
